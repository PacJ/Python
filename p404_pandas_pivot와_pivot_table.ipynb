#### pivot
- 각 컬럼의 값을 교차테이블 구성요소로 전달, 교차테이블 완성
- index, columns, values 컬럼을 각각 전달
- 요약기능 없음
- dataframe.pivot(index, columns,values)

```python
krecom_pivot = krecom_data.pivot(index='UserID', columns='MovieID', values='Rating')
krecom_pivot
```

```python
# inplace = True는 실제 데이터에 변경사항이 적용된다.
#krecom_pivot.fillna(method='ffil') # method='ffil' or method='pad' # 결측값을 앞방향으로 채운다.
#krecom_pivot.fillna(method='bfill') # method='bfill' or method='backfill' # 결측값을 뒷방향으로 채운다.
#krecom_pivot.fillna(krecom_pivot.mean() , axis=1)
krecom_pivot.fillna(0, inplace=True) #결측값을 0으로 채운다.
```

```python
krecom_pivot
```

#### pivot_table
- pivot기능과 유사, 더 많은 옵션 사용 가능
- index, columns, values컬럼을 각각 전달
- aggfunc옵션을 사용하여 요약 기능 전달 가능(기본은 평균)
- fill_value 옵션을 사용하여 NA값 대체 가능
- dataframe.pivot_table(index, columns, values, aggfunc='mean', fill_value)

```python
krecom_pivot_table = krecom_data.pivot_table(index='UserID', columns='MovieID', values='Rating', aggfunc='min', fill_value=0)
krecom_pivot_table
```

```python
df = pd.DataFrame(data= {'A':['foo', 'foo', 'foo', 'foo', 'foo', 'bar', 'bar', 'bar', 'bar'],
                         'B':['one', 'one', 'one', 'two', 'two', 'one', 'one', 'two', 'tow'],
                         'C':['small', 'large', 'large', 'small', 'small', 'large', 'small', 'small', 'large'],
                         'D':[1, 2, 2, 3, 3, 4, 5, 6, 7],
                         'E':[2, 4, 5, 5, 6, 6, 8, 9, 9]})
print(df)
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'])
table
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'], aggfunc={'D':np.mean,
                                                                         'E':[min, max, np.mean]})
table
```

#### stack
- unstack은 index를 column으로 만들어 준다.
- stack은 column을 index로 만들어 준다.
https://www.w3resource.com/pandas/dataframe/dataframe-stack.php

```python
df_single_level_no_index = pd.DataFrame([[0,2],[3,4]],
                                        columns = ['weight', 'height'])
df_single_level_no_index
```

```python
df_single_level_no_index.stack()
```

```python
df_single_level_no_index.stack().unstack()
```

```python
multicol1 = pd.MultiIndex.from_tuples([('weight', 'kg'),
                                       ('weight', 'pounds')])
df_multi_level_cols1 = pd.DataFrame([[3,4], [4,5]],
                                    index=['deer', 'monkey'],
                                    columns=multicol1)

df_multi_level_cols1
```

```python
df_multi_level_cols1.stack()
```

```python
df_multi_level_cols1.stack(0).unstack()
```

```python
swap = df_multi_level_cols1.stack(0).unstack()
swap.swaplevel(0,1, axis=1) # column의 계층(level) 순서변경
```
