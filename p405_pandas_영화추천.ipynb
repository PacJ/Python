#### merge와 join 비교

```python
df = pd.DataFrame({'key':['K0','K1','K2','K3','K4','K5'],
                   'A':['A0','A1','A2','A3','A4','A5']})
df
```

```python
other = pd.DataFrame({'key':['K0', 'K1', 'K2'],
                      'B':['B0','B1','B2']})
other
```

```python
# join()은 how속성의 값이 left이므로 left outer join을 실행한다.
#outer join
df.join(other.set_index('key'), on='key')
```

```python
# how=left
# Left Outer Join
other.join(df.set_index('key'), on='key')
```

```python
# how속성의 기본값이 inner이다
# inner join
df.merge(other)
```
